```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
next(data)
data = list(data)

name = ' '
fn=0
cn=0
close_name = ""
close_result = 0
far_name=""
far_result=0


for row in data :
    if name in row[1] :
        home = np.array(row[4:], dtype=int) 


for row in data :
    away = np.array(row[4:], dtype=int)  
    far= np.array(row[4:], dtype=int)   
    s = np.sum((home-away)**2)  
    f=np.sum((home-far)**2)
    if  s<fn and name not in row[1]:
        fn=s
        close_name=row[1]
        close_result=away
    elif f>fn and name not in row[1]:
        fn=f
        far_name=row[1]
        far_result=far  
        
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title(name+'        ')
plt.plot(home, label=name)
plt.plot(close_result, label=(close_name,"(",np.std(away),")"))
plt.plot(far_result,label=(far_name,"(",np.std(far),")"))
plt.legend()
plt.show()
```
