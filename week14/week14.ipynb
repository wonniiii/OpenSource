```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
next(data)
data = list(data)

name = '서울특별시 광진구'
fn=0
cn=999999999
close_name = ""
close_result = 0
far_name=""
far_result=0


for row in data :
    if name in row[1] :
        home = np.array(row[4:], dtype=int) 


for row in data :
    away = np.array(row[4:], dtype=int)  
    far= np.array(row[4:], dtype=int)   
    s = np.sum((home-away)**2)  
    f=np.sum((home-far)**2)
    if  s<cn and name not in row[1]:
        cn=s
        close_name=row[1]
        close_result=away
    elif f>fn and name not in row[1]:
        fn=f
        far_name=row[1]
        far_result=far  
        
close_sd=close_result[0]/101
far_sd=far_result[0]/101

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 지역과 가장 비슷하거나 다른 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(close_result, label="{0}(표준편차:{1:.3f})".format(close_name,close_sd))
plt.plot(far_result,label="{0}(표준편차:{1:.3f})".format(far_name,far_sd))
plt.legend()
plt.show()
```
